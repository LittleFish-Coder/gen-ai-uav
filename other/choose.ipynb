## Print the file names that in trainB but not in trainB_RIVER

```python
import os
import sys

def get_file_names(path):
    file_names = []
    for root, dirs, files in os.walk(path):
        for file in files:
            file_names.append(file)
    return file_names

def main(trainB, trainB_RIVER):
    trainB_files = get_file_names(trainB)
    trainB_RIVER_files = get_file_names(trainB_RIVER)
    for file in trainB_files:
        if file not in trainB_RIVER_files:
            print(file)

main('trainB', 'trainB_RIVER')
```

```python
import os

print(f"trainA Size: {len(os.listdir('trainA'))}")
print(f"trainB_RIVER Size: {len(os.listdir('trainB_RIVER'))}")

def main():
    trainA_dir = 'trainA'
    trainB_RIVER_dir = 'trainB_RIVER'

    # 从 trainA_ROAD 目录获取所有 JPG 图片的文件名（不包括扩展名）
    trainB_RIVER_files = {os.path.splitext(file)[0] for file in os.listdir(trainB_RIVER_dir) if file.endswith('.jpg')}

    # 遍历 trainA 目录中的 PNG 图片
    for file in os.listdir(trainA_dir):
        if file.endswith('.png'):
            # 提取不包括扩展名的文件名
            file_base_name = os.path.splitext(file)[0]
            if file_base_name not in trainB_RIVER_files:
                # 不在 trainA_ROAD 中的 PNG 文件，准备删除
                file_path = os.path.join(trainA_dir, file)
                os.remove(file_path)  # 删除文件
                print(f"Deleted: {file_path}")

if __name__ == "__main__":
    main()


print(f"trainA Size: {len(os.listdir('trainA'))}")
print(f"trainB_RIVER Size: {len(os.listdir('trainB_RIVER'))}")
```
